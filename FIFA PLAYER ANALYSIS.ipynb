```python
X = data[['Age','Potential','GKHandling','Penalties','Composure','Marking','StandingTackle','SlidingTackle','GKPositioning','GKReflexes']]
y = data['Overall']
y.shape
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
X_train
```

# Simple Linear Regression
Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_regressor=LinearRegression()
lin_regressor.fit(X_train,y_train)
```

```python
prediction_linear=lin_regressor.predict(X_test)
prediction_linear
```

```python
acc_LR=lin_regressor.score(X_test,y_test)
print(acc_LR)
```

```python
lin_regressor.score(X_train,y_train)
```

```python
print("Linear")
df = pd.DataFrame({'Actual': [y_test], 'Predicted':[prediction_linear]})
df
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
lasso_regressor=Lasso()
lasso_regressor.fit(X_train,y_train)
```

```python
prediction_lasso=lasso_regressor.predict(X_test)
prediction_lasso
```

```python
lasso_regressor.score(X_test,y_test)
```

```python
print("Lasso")
df = pd.DataFrame({'Actual': [y_test], 'Predicted':[prediction_lasso]})
df
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge_regressor=Ridge()
ridge_regressor.fit(X_train,y_train)
```

```python
prediction_ridge=ridge_regressor.predict(X_test)
prediction_ridge
```

```python
print("Rigde")
df = pd.DataFrame({'Actual': [y_test], 'Predicted':[prediction_ridge]})
df
```

```python
ridge_regressor.score(X_test,y_test)
```
